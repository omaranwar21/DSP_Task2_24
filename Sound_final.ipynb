# Parsel_mouth

```python
import parselmouth

sound = parselmouth.Sound("4_b.wav")
```

```python
from IPython.display import Audio

Audio(data=sound.values, rate=sound.sampling_frequency)
```

```python
from parselmouth.praat import call

manipulation = call(sound, "To Manipulation", 0.01, 75, 600)
```

```python
type(manipulation)
```

```python
manipulation.class_name
```

```python
pitch_tier = call(manipulation, "Extract pitch tier")

call(pitch_tier, "Multiply frequencies", sound.xmin, sound.xmax, 2)

call([pitch_tier, manipulation], "Replace pitch tier")
sound_octave_up = call(manipulation, "Get resynthesis (overlap-add)")
```

```python
type(sound_octave_up)
```

```python
Audio(data=sound_octave_up.values, rate=sound_octave_up.sampling_frequency)
```

# converted_to_female

```python
import librosa
import soundfile

y, sr = librosa.load("4_b.wav")
steps = 13.0
new_y = librosa.effects.pitch_shift(y, sr, steps)
```

```python
soundfile.write("pitchShifted.wav", new_y, sr)
```

```python
import IPython.display as ipd
from IPython.display import Audio
```

```python
ipd.Audio("pitchShifted.wav")
```

# Return_to_male

```python
y_conv, sr_conv = librosa.load("pitchShifted.wav")
steps2 = -13.0
new_y = librosa.effects.pitch_shift(y_conv, sr_conv, steps2)
```

```python
soundfile.write("converted.wav", new_y, sr_conv)
```

```python
ipd.Audio("converted.wav")
```
